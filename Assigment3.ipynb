Install libraries & load dataset

```python
# Step 1: Setup
!pip install --quiet kagglehub seaborn matplotlib

import kagglehub
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Download Iris dataset
path = kagglehub.dataset_download("uciml/iris")
print("Path to dataset files:", path)

# Load dataset
df = pd.read_csv(path + "/Iris.csv")

# Drop unnecessary column (Id)
if 'Id' in df.columns:
    df = df.drop(columns=['Id'])

df.head()
```

Overview of dataset


```python
print("Shape:", df.shape)
print("Columns:", list(df.columns))
df.info()
```

Visualization 1 – Histogram (Feature Distribution)


```python

plt.figure(figsize=(6,4))
plt.hist(df['SepalLengthCm'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

```

Visualization 2 – Scatter Plot (Feature Relationship)


```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='SepalLengthCm', y='PetalLengthCm', hue='Species', palette='Set2')
plt.title('Sepal vs Petal Length by Species')
plt.show()
```

Visualization 3 – Box Plot (Outlier Detection)

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Species', y='PetalWidthCm', palette='pastel')
plt.title('Box Plot of Petal Width by Species')
plt.show()
```

Visualization 4 – Pair Plot (Feature Relationships Overview)

```python
sns.pairplot(df, hue='Species', diag_kind='hist', palette='husl')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

Step 7: Visualization 5 – Correlation Heatmap

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```
