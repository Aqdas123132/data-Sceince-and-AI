```python
import pandas as pd

# Load the CSV file (case-sensitive name!)
df = pd.read_csv("/kaggle/input/iris/Iris.csv")

# Show first few rows
df.head(10)
```

Before cleaning step


```python
print("Shape:", df.shape)
print("\nMissing values per column:")
print(df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())

print("\nBasic statistics:")
df.describe()
```

Cleaning steps


```python
# Make a copy
df_clean = df.copy()

# 1) Remove duplicates
df_clean = df_clean.drop_duplicates()

# 2) Handle missing values (if any)
df_clean = df_clean.fillna(df_clean.median(numeric_only=True))

# 3) Remove 'Id' column if not needed
if 'Id' in df_clean.columns:
    df_clean = df_clean.drop(columns=['Id'])
```

After cleaning


```python
print("Shape after cleaning:", df_clean.shape)
print("\nMissing values after cleaning:")
print(df_clean.isnull().sum())

print("\nDuplicate rows after cleaning:")
print(df_clean.duplicated().sum())

df_clean.head(10)
```

Compare â€œBefore vs After


```python

print("**Before Cleaning**")
print("- Rows, Columns: 150 x 6")
print("- Duplicates: 0")
print("- Missing Values: 0")
print("- Outliers: None visible\n")

print("**After Cleaning**")
print("- Rows, Columns: 150 x 5 (removed Id column)")
print("- Duplicates: 0")
print("- Missing Values: 0\n")

print("âœ… Dataset ready for further analysis.")
```

Save the cleaned dataset


```python
df_clean.to_csv("iris_cleaned.csv", index=False)
print("Cleaned dataset saved successfully!")
```
